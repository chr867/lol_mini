```python
# 챔피언 데이터
champion_url ="https://ddragon.leagueoflegends.com/cdn/13.5.1/data/ko_KR/champion.json"
champion_res = requests.get(champion_url).json()
champion_list=[]
for key in champion_res['data']:
    champion_list.append(key)
```

```python
champion_df=[]
for name in champion_list:
    lst=[]
    lst.append(champion_res['data'][name]['version'])
    lst.append(champion_res['data'][name]['id'])
    lst.append(champion_res['data'][name]['key'])
    lst.append(champion_res['data'][name]['name'])
    lst.append(champion_res['data'][name]['title'])
    lst.append(champion_res['data'][name]['blurb'])
    lst.append(champion_res['data'][name]['info'])
    lst.append(champion_res['data'][name]['image'])
    lst.append(champion_res['data'][name]['tags'])
    lst.append(champion_res['data'][name]['partype'])
    lst.append(champion_res['data'][name]['stats'])
    champion_df.append(lst)
champion_df
```
